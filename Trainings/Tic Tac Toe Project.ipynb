* The board should be printed out every time a player makes a move
* You should be able to accept input of the player position and then place a symbol on the board


```python
def to_print_values(values):
    to_print = ""
    if len(values) == 0:
        pass
    else:
        to_print = "("
        for i in range(0, len(values)):
            to_print += str(values[i])
            if i == len(values) - 1:
                to_print += ")"
            elif i == len(values) - 2:
                to_print += " or "
            else:
                to_print += ", "

    return to_print
```

```python
def choose_icon(acceptable_values):

    icon = ''
    while icon not in acceptable_values:
        icon = input(f'Please enter your user icon: {to_print_values(acceptable_values)}')

        if icon.upper() not in acceptable_values:
            print(f'Invalid input. Please try again. Select {to_print_values(acceptable_values)}')

    return icon
```

```python
def print_board(current_board):

    board_option = {
        0: "   |   |   ",
        1: " {input1} | {input2} | {input3} ",
        2: "-----------"
    }
    print("\n--------- CURRENT BOARD ---------")
    for key in range(0, len(current_board)):
        board_value = current_board[key]

        if key == 1:
            print(board_option.get(2))

        print(board_option.get(0))
        print(board_option.get(1).format(input1=board_value[0], input2=board_value[1], input3=board_value[2]))
        print(board_option.get(0))

        if key == 1:
            print(board_option.get(2))
    print("---------------------------------\n")
```

```python

def print_available_choices(available_pos):
    board_option = {
            0: "   |   |   ",
            1: " {input1} | {input2} | {input3} ",
            2: "-----------"
        }

    for key in range(0, len(available_pos)):
        value = available_pos[key]

        if key == 1:
            print(board_option.get(2))

        print(board_option.get(0))
        print(board_option.get(1).format(input1=value[0], input2=value[1], input3=value[2]))
        print(board_option.get(0))

        if key == 1:
            print(board_option.get(2))
```

```python
def user_game(new_pos, user_symbol, current_board, available_pos):
    print("\nINSIDE THE GAME --------------------------------------")

    invalid_play = False
    has_row_match = 0
    has_column_match = 0
    has_diagonal_match = 0

    print(f'User played {user_symbol} in position {new_pos}')

    #while not valid_play:
    for key in range(0, len(current_board)):
        board_value = current_board[key]

        for val in range(0, len(board_value)) :
            value = board_value[val]
            #print(f'\n\t[{key}][{val}] == {board_value} : {value}')
            if available_pos[key][val] == new_pos :

                if value != ' ':
                    print(f'This position is not available anymore.\nPlease try again. Select {print_available_choices(available_pos)}')
                    invalid_play = True
                    break
                else:
                    board_value[val] = user_symbol
                    available_pos[key][val] = '-'

                    #Validate row
                    if val == 0 :
                        if current_board[key][1] == user_symbol : has_row_match += 1
                        if current_board[key][2] == user_symbol : has_row_match += 1
                    if val == 1 :
                        if current_board[key][0] == user_symbol : has_row_match += 1
                        if current_board[key][2] == user_symbol : has_row_match += 1
                    if val == 2 :
                        if current_board[key][0] == user_symbol : has_row_match += 1
                        if current_board[key][1] == user_symbol : has_row_match += 1

                    #Validate column
                    if key == 0:
                        if current_board[1][val] == user_symbol : has_column_match += 1
                        if current_board[2][val] == user_symbol : has_column_match += 1
                    if key == 1 :
                        if current_board[0][val] == user_symbol : has_column_match += 1
                        if current_board[2][val] == user_symbol : has_column_match += 1
                    if key == 2 :
                        if current_board[0][val] == user_symbol : has_column_match += 1
                        if current_board[1][val] == user_symbol : has_column_match += 1

                    #validate diagonal
                    if key == 0 and val == 0 :
                        if current_board[1][1] == user_symbol : has_diagonal_match += 1
                        if current_board[2][2] == user_symbol : has_diagonal_match += 1
                    if key == 0 and val == 2:
                        if current_board[1][1] == user_symbol : has_diagonal_match += 1
                        if current_board[2][0] == user_symbol : has_diagonal_match += 1

                    if key == 1 and val == 1:
                        if current_board[0][0] == user_symbol and current_board[2][2] == user_symbol : has_diagonal_match += 2
                        if current_board[0][0] == user_symbol and current_board[2][2] == ' ' : has_diagonal_match += 1

                        if current_board[0][2] == user_symbol and current_board[2][0] == user_symbol : has_diagonal_match += 2
                        if current_board[0][2] == user_symbol and current_board[2][0] == ' ' : has_diagonal_match += 1

                    if key == 2 and val == 0 :
                        if current_board[1][1] == user_symbol : has_diagonal_match += 1
                        if current_board[0][2] == user_symbol : has_diagonal_match += 1
                    if key == 2 and val == 2:
                        if current_board[0][0] == user_symbol : has_diagonal_match += 1
                        if current_board[1][1] == user_symbol : has_diagonal_match += 1
        if invalid_play:
            break

    game_is_over = (has_row_match == 2 or has_column_match == 2 or has_diagonal_match == 2)
    print("------------------------------------------------------\n")
    return {'game_is_over': game_is_over, 'invalid_play': invalid_play, 'current_board': current_board, 'available_pos': available_pos}
```

```python
def position_choice(available_positions):
    choice = ""
    good_choice = any(choice in x for x in acceptable_positions)

    while not good_choice:
        choice = input(f'Pick a position to replace {available_positions}')
        good_choice = any(choice in x for x in acceptable_positions)
        if not good_choice:
            #clear_output()
            print(f'Sorry, but you did not choose a valid position {print_available_choices(available_positions)}')

    #clear_output()
    return choice
```

```python
from random import randint

print("Welcome to Tic Tac Toe!")

acceptable_user_icons = ['X', 'O']
acceptable_positions = [['7', '8', '9'], ['4', '5', '6'], ['1', '2', '3']]
game_board = [[' ',' ',' '], [' ',' ',' '], [' ',' ',' ']]

# The first user should select the icon to play
user1_icon = choose_icon(acceptable_user_icons)
acceptable_user_icons.remove(user1_icon)
# The second user will play with the other icon
user2_icon = acceptable_user_icons[0]
print(f'User 1 has selected {user1_icon}. User 2 will play with {user2_icon}')

keep_playing = True
game_over = False

users = {0: user1_icon, 1: user2_icon}

user_selected = randint(0,1)
print(f'LETS START !!')
num_plays = 0

#while keep_playing or not game_over:
while keep_playing:
    num_plays += 1

    user_icon = users.get(user_selected)
    print(f'User {user_selected} play turn #{num_plays} with icon {user_icon}!!')

    selected_position = position_choice(acceptable_positions)
    results = user_game(selected_position, user_icon, game_board, acceptable_positions)

    game_board = results['current_board']
    acceptable_positions = results['available_pos']
    keep_playing = not results['invalid_play']

    print_board(game_board)
    if results['game_is_over']:
        print(f'Congratulations {user_selected}! You won!')
        break

    # Change user
    if user_selected == 1 :
        user_selected = 0
    else:
        user_selected = 1
```
